Breath-First Search


```python
def bfs(start, goal):
    """
    BFS using list as queue (FIFO).
    """
    start_idx = node_to_idx(start)
    goal_idx = node_to_idx(goal)
    
    visited = np.zeros(n, dtype=bool)
    queue = []
    queue.append((start_idx, [start_idx]))

    while len(queue) > 0:
        current, path = queue.pop(0)
        
        if current == goal_idx:
            # sum edge weights along the path
            path_cost = sum(graph[path[i], path[i+1]] for i in range(len(path)-1))

            return [idx_to_node(i) for i in path], path_cost
        
        if not visited[current]:
            visited[current] = True
            
            for neighbor in range(n):
                if graph[current, neighbor] != INF and not visited[neighbor]:
                    queue.append((neighbor, path + [neighbor]))
    
    return None, None

# Arad to Bucharest 
start_node = 'Arad'
goal_node = 'Bucharest'
path, total_cost = bfs(start_node, goal_node)
if path:
    print(f"BFS: {' -> '.join(path)}")
    print(f"Total Cost: {total_cost}")
else:
    print("BFS: No path found")
```

Depth-First Search

```python
def dfs(start, goal):
    """
    DFS using list as stack (LIFO).
    """
    start_idx = node_to_idx(start)
    goal_idx = node_to_idx(goal)
    
    visited = np.zeros(n, dtype=bool)
    stack = []
    stack.append((start_idx, [start_idx]))
    
    while len(stack) > 0:
        current, path = stack.pop()
        
        if current == goal_idx:
            # sum edge weights along the path
            path_cost = sum(graph[path[i], path[i+1]] for i in range(len(path)-1))

            return [idx_to_node(i) for i in path], path_cost
        
        if not visited[current]:
            visited[current] = True
            
            for neighbor in range(n):
                if graph[current, neighbor] != INF and not visited[neighbor]:
                    stack.append((neighbor, path + [neighbor]))
    
    return None, None

# Arad to Bucharest 
start_node = 'Arad'
goal_node = 'Bucharest'
path, total_cost = dfs(start_node, goal_node)
if path:
    print(f"DFS: {' -> '.join(path)}")
    print(f"Total Cost: {total_cost}")
else:
    print("DFS: No path found")
```

Uniform Cost Search

```python
def ucs(start, goal):
    """
    UCS using priority queue (min-cost selection).
    """
    start_idx = node_to_idx(start)
    goal_idx = node_to_idx(goal)
    
    visited = np.zeros(n, dtype=bool)
    priority_queue = []
    priority_queue.append([0, start_idx, [start_idx]])
    
    while len(priority_queue) > 0:
        # Find minimum cost entry
        min_idx = 0
        for i in range(1, len(priority_queue)):
            if priority_queue[i][0] < priority_queue[min_idx][0]:
                min_idx = i
        
        cost, current, path = priority_queue.pop(min_idx)
        
        if current == goal_idx:
            return [idx_to_node(i) for i in path], cost
        
        if not visited[current]:
            visited[current] = True
            
            for neighbor in range(n):
                if graph[current, neighbor] != INF and not visited[neighbor]:
                    new_cost = cost + graph[current, neighbor]
                    new_path = path + [neighbor]
                    priority_queue.append([new_cost, neighbor, new_path])
    
    return None, None

# Arad to Bucharest 
start_node = 'Arad'
goal_node = 'Bucharest'
path, total_cost = ucs(start_node, goal_node)
if path:
    print(f"UCS: {' -> '.join(path)}")
    print(f"Total Cost: {total_cost}")
else:
    print("UCS: No path found")
```
